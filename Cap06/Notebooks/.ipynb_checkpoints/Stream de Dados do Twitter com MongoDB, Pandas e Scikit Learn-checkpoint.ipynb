# Análise de Dados com Pandas e SciKit-Learn

```python
# Criando um dataset com dados do mongo e usando list Comprehension
dataset = [{'created_at': item['created_at'], 'text': item['text'],} for item in col.find()]
```

```python
!pip install pandas
```

```python
# Importando o módulo Pandas
import pandas as pd
```

```python
# Criando um objeto do tipo DataFrame
df = pd.DataFrame(dataset)
```

```python
# Imprimindo o DataFrame
df
```

```python
!pip install sklearn
```

```python
# Importamos o módulo Scikit learn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Usamos o método CountVectorizer para criar uma matriz de documentos
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# Contando o número de ocorrências das principais palavras em nosso dataset
word_count = pd.DataFrame(cv.get_feature_names(), columns=['word'])
word_count['count'] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values('count', ascending=False).reset_index(drop=True)
word_count[:50]
```
